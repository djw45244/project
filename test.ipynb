# Example of a simple Binder

This Jupyter notebook uses a Python environment with a few libraries, all of which were specificied using a conda environment.yml file

#### Set up plotting

```python
import numpy
import pandas
import matplotlib.pyplot
%matplotlib inline
```

#### Generate fake data

```python
df = pandas.DataFrame(numpy.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))
```

#### Compute and plot a cumulative sum

```python
fig,ax = matplotlib.pyplot.subplots()
ax.plot(df.cumsum())
ax.set_xlabel('Index')
ax.set_ylabel('Cumulative sum')
ax.legend(labels=df.columns.to_list())
```
